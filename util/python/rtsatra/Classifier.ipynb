```python
from sklearn.pipeline import Pipeline
from sklearn.cross_validation import KFold, ShuffleSplit, StratifiedShuffleSplit
from sklearn.grid_search import GridSearchCV
from sklearn.svm import SVR, SVC
from sklearn.ensemble import ExtraTreesClassifier, RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import explained_variance_score, r2_score, f1_score, matthews_corrcoef
from IPython.display import display
```

```python
import nibabel as nb
```

```python
filename = '../merged.nii.gz'
data = nb.load(filename).get_data()
print data.shape
```

```python
from nipype.interfaces.fsl import BET
```

```python
res = BET(in_file=filename, mask=True).run()
```

```python
meanimg = nb.load(res.outputs.out_file).get_data()
print meanimg.shape
```

```python
mask = nb.load(res.outputs.mask_file).get_data().astype(int)
print mask.shape
```

```python
Xall = data[mask>0, :]
Xall.shape
```

```python
X = Xall[:, :64].T
Y = np.hstack((np.ones(16), -1*np.ones(16), np.ones(16), -1*np.ones(16)))
print X.shape
print Y.shape
```

```python
results = [[], []]
scores = []
clfs = []
for train, test in StratifiedShuffleSplit(Y, n_iter=200, test_size=0.2):
    if 1:
        clf = ExtraTreesClassifier(n_estimators=100, max_depth=None, min_samples_split=1)
    else:
        clf = Pipeline([('scaler', StandardScaler()),
                            ('model', SVC())])
        parameters = {'model__kernel': ('linear', 'rbf',), 
                          'model__C':[0.1, 1, 10, 100]}
        clf = GridSearchCV(clf, parameters, n_jobs=8)
    ytest = clf.fit(X[train], Y[train]).predict(X[test])
    results[0].extend(Y[test])
    results[1].extend(ytest)
    scores.append(f1_score(Y[test], ytest, [-1, 1]))
    clfs.append(clf.feature_importances_.tolist())
score = f1_score(results[0], results[1], [-1,1])
```

```python
hist(scores, 10)
print score
```

```python
features = np.array(clfs).mean(axis=0)
```

```python
featimg = np.zeros(mask.shape)
featimg[mask>0] = features
print featimg.shape
featimg = featimg*(featimg>np.percentile(features, 90))
print featimg.shape
featimg[featimg<=0] = np.nan
print featimg.shape
```
